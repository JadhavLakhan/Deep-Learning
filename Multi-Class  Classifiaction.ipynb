```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
df.head()
```

```python
le={}
for i in df.select_dtypes(include='object').columns:
    le[i]=LabelEncoder()
    df[i]=le[i].fit_transform(df[i])
```

```python
df.head()
```

```python
X=df.drop("Drug",axis=1)
y=df["Drug"]
```

```python
std=StandardScaler()
X_trns=std.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_trns,y,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(32, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(16, activation='relu'),
    Dense(5, activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=50,validation_split=0.1,batch_size=16)
```

Setting **epochs=50** means the model will iterate over the entire training data 50 times to learn and adjust the weights.

In this case, the model will use 16 samples per batch. After processing 16 samples, the model will update the weights (i.e., the model will "learn" from those 16 samples).

Validation Data: The remaining $10%$ is used to evaluate the model's performance during each epoch (iteration) of training. It allows you to monitor the model's performance on unseen data, which helps you assess how well it generalizes. This helps in detecting issues like overfitting.

**Example:**
Suppose you have 1000 training samples:

With **validation_split=0.1** , 100 samples (10% of 1000) are set aside for validation.

The remaining 900 samples are used to train the model.

The model will train on those 900 samples and, after each epoch, will evaluate its performance using the 100 validation samples.

```python
loss,accuracy=model.evaluate(X_test,y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

The model is achieving good performance with a test accuracy of 95%. This means that, out of 100 test samples, the model is correctly predicting the class of 95 of them.

**Loss of 0.2225** indicates that the predictions are not perfect but relatively close to the true labels.

Depending on the complexity of the dataset and problem, **95%** accuracy can be considered very good. However, whether it's satisfactory depends on the specific context and the baseline performance you are comparing it to.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
intermediate_layer_model = Sequential(model.layers[:-1])  # Remove last layer
X_transformed = intermediate_layer_model.predict(X_trns)
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(X_transformed, y)
```

```python
y_pred = tree_model.predict(X_transformed)
print("Hybrid Model Accuracy:", accuracy_score(y, y_pred))
```
