## Step I:Load Your Dataset

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
import seaborn as sn
```

```python
df=pd.read_csv(r"D:\Downloads\house_price_regression_dataset.csv")
data=fetch_california_housing()
```

```python
data.target_names
```

```python
X=pd.DataFrame(data=data,columns=data.feature_names)
y=pd.DataFrame(data=data.target,columns=data.target_names)
```

```python
X.shape,y.shape
```

## Step II: Split the Data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X=df.drop(columns='House_Price',axis=1)
y=df['House_Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

## Step II: Scale the Features

```python
std=StandardScaler()
X_train=std.fit_transform(X_train)
X_test=std.transform(X_test)
```

## Step IV: Build the Neural Network Model

```python
from keras.models import Sequential
from keras.layers import Dense
from tensorflow import keras
from tensorflow.keras import layers
```

```python
Z=X_train.shape[1],
```

```python
Z
```

If the problem is more complex, you can experiment with two or three hidden layers

```python
model=Sequential([
    layers.Input(shape=Z),
    layers.Dense(10,activation='relu'), #ReLU (Rectified Linear Unit)
    layers.Dense(1)
])
```

```python
model.compile(optimizer='adam' , loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=1)
```

```python
model = keras.Sequential([
    layers.Input(shape=(X_train.shape[1],)),
    layers.Dense(10, activation='relu'),  # Hidden layer with 10 neurons
    layers.Dense(1)  # Output layer
])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=1)
#You can set it to 0 for no output, 1 for a progress bar, and 2 for one line per epoch.
```

```python
loss=model.evaluate(X_test,y_test)
print("loss:-",loss)
```

```python
predictions = model.predict(X_test)
print(predictions)
```

```python
# Display some predictions
for i in range(5):
    print(f'Predicted: {predictions[i][0]}, Actual: {y_test.values[i]}')
```

```python
weights, biases = model.layers[0].get_weights()
print(f'Weights (slope): {weights[0][0]}')
print(f'Biases (intercept): {biases[0]}')
```
